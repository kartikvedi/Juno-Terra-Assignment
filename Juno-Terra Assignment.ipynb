```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Untitled 1.csv')
```

```python
data
```

```python
data.drop(['ID','Ozone','Methane'],1,inplace=True)
```

```python
data.info()
```

As we see here there is only 1 city which is emitting the Methane gas i.e Bejing whose value is 1840.2364

```python
plt.figure(figsize=(10,4))
sns.barplot(x='City',y="NO",data=data)
plt.xticks(rotation=45)
plt.show()
```

In the above figure we can see that the maximum Nitrogendioxide emission is done in Moscow city.

```python
plt.figure(figsize=(10,4))
sns.barplot(x='City',y="CO",data=data)
plt.xticks(rotation=45)
plt.show()
```

Here Shanghai is the major emitter of Carbon Monoxide.

In the emisson of Carbon monoxide,followings are the data in the descending order,which indiacates that the shanghai is ranked 1 for emitting most of the Carbon Monoxide gas compairing with others.

```python
data.sort_values(by='CO',ascending=False,)
```

For the Nitrogen dioxide,Moscow is ranked 1 for emitting the gas compairing with others.
